```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("homeprices.csv")
data
```

```python
X=data['area'].values
y=data['price'].values

meanx=np.mean(X)
meany=np.mean(y)
print("mean of x=",meanx)
print("mean of y=",meany)
```

```python
m=len(X)
num=0
den=0
for i in range(m):
    num+=(X[i]-meanx)*(y[i]-meany)
    den+=(X[i]-meanx)**2
m=num/den
c=meany-(m*meanx)
print("Slope=",m)
print("intercept=",c)
```

```python
reg_line=m*X+c
plt.scatter(X,y,color='blue',label="data Points")
plt.plot(X,reg_line,color='red',label='Regression line')
plt.xlabel('area')
plt.ylabel('price')
plt.legend()
plt.show()
```

```python
# calculating R2 Score for checking the efficency of models
m=len(X)
num=0
den=0
for i in range(m):
    y_pred=c+m*X[i]
    num+=(y[i]-meany) ** 2
    den+=(y[i]-y_pred) ** 2
    r2=1-num/den
print(r2)
```
