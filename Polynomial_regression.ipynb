```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("home.csv")
data
```

```python
X = data['area'].values.reshape(-1, 1)
y = data['price'].values
```

```python
degree = 3
X_poly = np.ones((len(X), degree + 1))
for d in range(1, degree + 1):
    X_poly[:, d] = X[:, 0] ** d
coefficients = np.linalg.inv(X_poly.T.dot(X_poly)).dot(X_poly.T).dot(y)
y_pred = X_poly.dot(coefficients)
rmse = np.sqrt(np.mean((y - y_pred) ** 2))
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
plt.scatter(X, y, color='blue', label="Data Points")
plt.plot(X, y_pred, color='red', label='Polynomial Regression')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Polynomial Regression (Degree {})'.format(degree))
plt.legend()
plt.show()
```
